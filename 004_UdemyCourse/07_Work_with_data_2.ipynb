3. Maping

```python
df = DataFrame({'city':['Alma','Brian Head', 'Fox Park'],
               'altitude':[3158, 3000, 2762]})
```

```python
df
```

```python
state_map = {'Alma':'Colorado', 'Brian Head': 'Utah', 'Fox Park':'Wyoming'}
```

```python
state_map
```

```python
df['state'] = df['city'].map(state_map)
```
