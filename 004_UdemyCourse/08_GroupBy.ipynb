```python
animals = DataFrame(np.arange(16).reshape(4,4),
                   columns=['A','B','C','D'],
                   index=['Dog', 'Cat', 'Bird', 'Mouse'])
```

```python
animals
```

```python
animals.ix[1:2,['A','C']] = np.nan
```

```python
animals
```

```python
 behavior_map = {'A':'good',
                'B':'bad',
                'C':'good',
                'D':'bad'}
```

```python
animals_with_map = animals.groupby(behavior_map, axis=1)
```
